```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pima = pd.read_csv("diabetes.csv")
starbucks = pd.read_csv("starbucks_drinkMenu_expanded.csv")
netflix = pd.read_csv ("netflix_titles.csv")
```

```python
print("PIMA Diabetes:\n", pima.head(), "\n")
print("Starbucks:\n", starbucks.head(), "\n")
print("Netflix:\n", netflix.head(), "\n")

print("PIMA Info:")
pima.info()
print("\nStarbucks Info:")
starbucks.info()
print("\nNetflix Info:")
netflix.info()

print("\nPIMA Describe:\n", pima.describe())
print("\nStarbucks Describe:\n", starbucks.describe())
print("\nNetflix Describe:\n", netflix.describe(include='all'))  # include='all' to show non-numeric too
```

```python
starbucks['Calories'] = pd.to_numeric(starbucks['Calories'], errors='coerce')
```

```python
# Part 2: Univariate Analysis , Starbucks Dataset

plt.figure(figsize= (6,4))
plt.hist(starbucks['Calories'].dropna(), bins=15, color='lightblue', edgecolor='Black')
plt.title('Distribution Of Calories In Starbucks Drinks')
plt.xlabel('Calories')
plt.ylabel('Number of Drinks')
plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```

```python


plt.figure (figsize=(6,4))
sns.boxplot (x = starbucks ['Calories'], color='lightpink')
plt.title ('Box Plot Of Calories In Starbucks Drinks')
plt.xlabel ('Calories')
plt.show()
```

Interpretation: 

The histogram actually shows the entire spread and frequency of various calorie values. However, on the other hand, the boxplot may show the median calorie amount and any outliers present. Majority of the the drinks fall below 200 calories. Overall, the data is right-skewed becuase there are very few high-calorie drinks.

```python
# Part 2: Univariate Analysis , PIMA Dataset


plt.figure (figsize=(6,4))
sns.kdeplot (data=pima, x='Age', fill=True, color='lightgreen')
plt.title ('Age Distribition in PIMA Diabetes Dataset (KDE)')
plt.xlabel ('Age')
plt.ylabel ('Density')
plt.grid (True)
plt.show()
```

```python
# Part 2: Univariate Analysis , PIMA Dataset


plt.figure (figsize=(6,4))
sns.histplot (data=pima, x='Age', kde =True, bins=15, color='skyblue', edgecolor='grey')
plt.title ('Age Distribution in PIMA Diabetes Dataset (Histogram, KDE)')
plt.xlabel ('Age')
plt.ylabel ('Frequency')
plt.grid (True)
plt.show ()
```

```python
# Part 3: Bivariate Analysis , Starbucks Dataset


plt.figure (figsize=(6,4))
sns.scatterplot(data=starbucks, x='Calories', y=' Sugars (g)', color='green', edgecolor='black')
plt.title ('Do more calories = more sugar in Starbucks Drinks?')
plt.xlabel ('Calories')
plt.ylabel ('Sugars (g)')
plt.grid (True)
plt.show ()
```

Interpretation:

Clearly, there is a positive correlation between calories and sugars (g) in Starbucks Drinks according to the dataset. As the calorie content in the drink increase, sugars content also increases. Thus, one may come to the conclusion that most of the Starbucks drinks actually get a huge portion of their calories from sugar.

```python
# Part 3: Bivariate Analysis , PIMA Dataset


plt.figure (figsize=(6,4))
sns.violinplot(data=pima, x='Outcome', y='Glucose', palette='pastel')
plt.title ('Glucose Levels vs Diabetes (Yes/No)')
plt.xlabel ('Diabtetes Outcome (0=No, 1=Yes)')
plt.ylabel ('Glucose Level')
plt.grid (True)
plt.show()
```

Interpretation:

People who have Diabetes (Outcome=1) usually have higher glucose levels than those who don't have Diabetes (Outcome=0). In the diabetic group of people, there is a wider distribution. The median glucose level is higher in the diabetic group as compared to the non-diabetic group.

```python
titles_per_year = netflix['release_year'].value_counts().sort_index()
```

```python
# Part 3: Bivariate Analysis , Netflix Dataset


plt.figure(figsize=(6,4))
titles_per_year.plot(kind='line', color='orange', marker='o')
plt.title('Number of Netflix Titles Released Every Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.show()
```

```python
movie_counts = netflix[netflix['type'] == 'Movie']['release_year'].value_counts().sort_values()
tv_counts = netflix[netflix['type'] == 'TV Show']['release_year'].value_counts().sort_values()

plt.figure(figsize=(12,6))
plt.plot(movie_counts.index, movie_counts, label='Movie', color='purple')  
plt.plot(tv_counts.index, tv_counts.values, label='TV Shows', color='blue') 

plt.title('Number of Netflix Releases by Year')  
plt.xlabel('Release Year')
plt.ylabel('Number of Titles Released')
plt.grid(True)
plt.legend(loc='lower right')  
plt.show()
```

```python
# Part 4: Multivariate Analysis and Customisation , PIMA Dataset HeatMap


selected = ['Glucose', 'BMI', 'Age', 'Insulin', 'BloodPressure']
corr_matrix = pima[selected].corr()

plt.figure(figsize=(7,5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Selected PIMA Features')
plt.tight_layout()
plt.show()
```

Interpretation:

Definitely, after considering and analysing this heatmap, a strong positive correlation can be seen between Glucose and BMI - thus, people with higher Glucose Levels have higher BMI. On the other hand, Insulin and Age show little to no correlation, which basically means that these two vary independently in this particular dataset.

```python
# PART 5: Subplots

fig, axs = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Exploring Data Across Datasets', fontsize=16)

axs[0, 0].hist(pima['BloodPressure'], bins=15, color='teal', edgecolor='black')
axs[0, 0].set_title('Blood Pressure Distribution (PIMA)')
axs[0, 0].set_xlabel('Blood Pressure')
axs[0, 0].set_ylabel('Frequency')

axs[0, 1].scatter(starbucks['Calories'], starbucks[' Protein (g) '], color='orchid')
axs[0, 1].set_title('Calories vs Protein in Starbucks Drinks')
axs[0, 1].set_xlabel('Calories')
axs[0, 1].set_ylabel('Protein(g)')

sns.countplot(x='Beverage_category', data=starbucks, ax=axs[1, 0], palette='pastel')
axs[1, 0].set_title('Beverage Category Count')
axs[1, 0].set_xlabel('Category')
axs[1, 0].set_ylabel('Count')
axs[1, 0].tick_params(axis='x', rotation=45)

sns.boxplot(x='Outcome', y='Age', data=pima, ax=axs[1, 1], palette='Set2')
axs[1, 1].set_title('Age by Diabetes Outcome')
axs[1, 1].set_xlabel('Diabetes Outcome')
axs[1, 1].set_ylabel('Age')
plt.tight layout()
plt.show()
```

```python
import geopandas as gpd
import os
import numpy as np

folder = '/Users/jia/Desktop/randoms/ashoka-horizons-data-science/week3/STATES/'
states = [
    'ANDAMAN & NICOBAR', 'ARUNACHAL PRADESH', 'ANDHRA PRADESH', '.DS_Store', 'LAKSHADWEEP', 'TELANGANA', 'MANIPUR', 'MADHYA PRADESH', 'DAMAN & DIU', 'HARYANA', 'GUJARAT', 'SIKKIM', 'HIMACHAL PRADESH', 'DELHI', 'GOA', 'PUNJAB', 'JAMMU & KASHMIR', 'KARNATAKA', 'MAHARASHTRA', 'DADRA & NAGAR HAVELI', 'ORISSA', 'RAJASTHAN', 'CHANDIGARH', 'KERALA', 'CHHATTISGARH', 'TRIPURA', 'MIZORAM', 'LADAKH', 'ASSAM', 'MEGHALAYA', 'TAMIL NADU', 'WEST BENGAL', 'UTTARAKHAND', 'UTTAR PRADESH', 'PUDUCHERRY', 'BIHAR', 'JHARKHAND', 'NAGALAND'
]
gdfs =[]
for state in states:
    path = os.path.join (folder, state, f"{state}.geojson")
    if os.path.exists (path):
        gdf = gpd.read_file (path)
        gdf["State"] = state
        gdfs.append (gdf)

combined_gdf = gpd.GeoDataFrame (pd.concat (gdfs, ignore_index=True))

combined_gdf ["Dummy Score"] = np.random.randint (10,100, size=len(combined_gdf))

combined_gdf.plot (column ='Dummy Score', cmap ='plasma', legend =True, figsize = (10,12), edgecolor ='black')
plt.title ('Chloropleth Map of Indian States (Dummy Scores)', fontsize=15)
plt.axis ('off')
plt.show()
```
